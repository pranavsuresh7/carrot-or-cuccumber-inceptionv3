```python
from keras.models import load_model
from keras.applications.inception_v3 import preprocess_input
from keras.preprocessing import image
```

```python
from PIL import Image
from matplotlib.pyplot import imshow
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
import os
import time
```

```python
from os import listdir
import tensorflow as tf
from os.path import isfile, join
import zipfile
```

```python
model = load_model('mycarrot.h5')
```

```python

cap = cv2.VideoCapture(0)
cap.set(cv2.CAP_PROP_FPS, 1);
current_frame = 0

while(True):
    # Capture frame-by-frame
    ret, frame = cap.read()
    # created a name for each images to be store
    name = "/home/pranav/Desktop/opencvimages/"+str(current_frame)+".jpg"
    print("passing the frames to the folder created",name)
    
    # Our operations on the frame come here
#     gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    cv2.imwrite(name,frame)
    imgd = np.array(frame)
    current_frame+=1
    cv2.imshow('name',imgd)
    # Display the resulting frame
    img=image.load_img(name,target_size=(224,224))
    x=image.img_to_array(img)
    x=preprocess_input(x).reshape(1,224,224,3)
    prediction=model.predict(x)
    if prediction[0,0]>0.5:
        print('cuccumber')
    else:
        print('carrot')
    time.sleep(5)
    if cv2.waitKey(1) & 0xFF == ord('k'):
        break

# When everything done, release the capture
cap.release()
cv2.destroyAllWindows()
```
